```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
```

```python
# Sample text to translate into tamil

text = "it is a piece of cake"
```

```python
model_inputs = tokenizer(text, return_tensors = "pt")
```

```python
model_inputs
```

```python
# Changing the language into french and using the code "fr_XX"
generated_tokens = model.generate(**model_inputs, forced_bos_token_id = tokenizer.lang_code_to_id ["ru_RU"])
translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
translation
```
